```python
import torch 
from torch import nn

import torchvision
from torchvision import datasets
from torchvision import transforms
from torchvision.transforms import ToTensor 

import matplotlib.pyplot as plt

print(torch.__version__)
print(torchvision.__version__)
```

```python
#Using Fashion MNIST Database

#IMG NET as well cool

train_data=datasets.FashionMNIST(root="data", #where it is downloaded
                                 train=1, #do we want training dataset
                                 download=1, #de we download it
                                 transform=ToTensor(), #do we transofrm data
                                 target_transform=None) #do we transform labels/targets

test_data=datasets.FashionMNIST(root="data", #where it is downloaded
                                 train=0, #do we want training dataset
                                 download=1, #de we download it
                                 transform=ToTensor(), #do we transofrm data
                                 target_transform=None) #do we transform labels/targets
```

```python
torch.manual_seed(42)

fig,_=plt.subplots(figsize=(9,9))
plt.axis("off")

rows,cols=4,4
for i in range(1,rows*cols+1):
    rand_idx=torch.randint(0,len(train_data),size=[1]).item()
    img,label=train_data[rand_idx]
    fig.add_subplot(rows,cols,i)
    plt.imshow(img.squeeze(),cmap="gray")
    plt.title(class_names[label])
    plt.axis("off");

plt.show()
    
```
